```python
from urllib.request import getproxies, urlopen
import re
import json
import datetime
import time
import pandas as pd

import tkinter as tk
import seaborn as sns
from PIL import Image, ImageTk
import math
%matplotlib inline
#輸入日期與股票
#輸出sql與[日期,開盤價,最高價,最低價,收盤價]共十二月份


# 自定義一個可以輸入股票與日期的函式
# input_stock=int(input("請輸入想要查找的股票代號"))
# input_date=int(input("請輸入想存取該股票之年分與月份(系統會自動抓取後十二個月分之資料`, 輸入範例:20200101)"))



def getPrice():
    stock = symbol_entry.get()
    date = date_entry.get()
   # stock=str(stock)
    #date=str(date)
    html_1 = urlopen("https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=json&date="+date+"&stockNo="+stock)
    content = html_1.read().decode('utf-8')
    jcontent = json.loads(content)
    data = jcontent['data']
    data = [[i[0], float(i[3].replace(',', '')), float(i[4].replace(',', '')), float(
        i[5].replace(',', '')), float(i[6].replace(',', ''))] for i in data]
    frame=pd.DataFrame(data,columns=['date','opening price','High Price','Low Price','closing price'])
    
    #繪圖
    pic=frame.plot(x='date',y='Low Price',title='low price run chart')
    img=pic.get_figure() #取得圖片
    img.savefig('figure.png')#存檔
    
    #顯示文字
    result = '證券代號{}\n交易日期{}\n{}\n'.format(stock,date,frame)
    result_label.configure(text=result)
    
    # 讀取圖片&轉換成Tkinter可以用的圖片
    img = Image.open('figure.png')  
    img =  ImageTk.PhotoImage(img)                        

    # 宣告標籤並且設定圖片
    figure = tk.Label(window, image=img)                   
    # 排版位置
    figure.image = img#避免被recycle
    figure.pack()
                   

 

window = tk.Tk()
window.title('Finance crawler')
window.geometry('800x600')
window.configure(background='white')

    
header_label = tk.Label(window, text='Stock Transaction Information')
header_label.pack()#排版

symbol_frame = tk.Frame(window)
symbol_frame.pack(side=tk.TOP)
symbol_label = tk.Label(symbol_frame, text='證券代號')
symbol_label.pack(side=tk.LEFT)
symbol_entry = tk.Entry(symbol_frame)
symbol_entry.pack(side=tk.LEFT)

date_label = tk.Label(symbol_frame, text='日期')
date_label.pack(side=tk.LEFT)
date_entry = tk.Entry(symbol_frame)
date_entry.pack(side=tk.LEFT)


search_btn = tk.Button(window, text='查詢', command=getPrice)
search_btn.pack()

result_label = tk.Label(window)
result_label.pack()


window.mainloop()
```
